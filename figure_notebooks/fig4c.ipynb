```python
urine_logcpmtmm = toolkit.logcpm_tmm("../data/20231228_filtered_sed_supt_urineONLY.csv",
                                      "../data/diffex/urine_only/20231228_urineOnly_TMM.csv")
all_samps = np.unique([i.split("_")[0] for i in urine_logcpmtmm.columns])
all_samps = [i + "supt" if i == "1754" else i for i in all_samps]

# read in metabo_data and preprocess
metabo_data = pd.read_excel("../data/SEVO001_CombineSubmit_July2023.xlsx",
                           sheet_name = 'Unique Annotated',
                            index_col = 0, header = 1)

metabo_data.columns = metabo_data.iloc[2, :]

metabo_data = metabo_data[metabo_data.INCHIKEY != 'Internal Standard']

metabo_data = metabo_data.iloc[3:, :]

missing = [i for i in metabo_data.INCHIKEY if '-' not in i]
missing += [i for i in metabo_data.INCHIKEY if '?' in i]

metabo_data = metabo_data[~metabo_data['INCHIKEY'].isin(missing)]

# fed metabo_data.INCHIKEY to classyfire

index_lst = ['Alignment ID', 'Average Rt(min)', 'Average Mz', 'Metabolite name', 'INCHIKEY']

metabo_data.set_index(index_lst,
                     inplace = True)

metabo_data = metabo_data.iloc[:, 3:]

samples = [i for i in metabo_data.columns if 'QC' not in i and 'BK' not in i]
metabo_data = metabo_data.loc[:, samples]

metabo_data.columns = [i.split("_")[2] for i in metabo_data.columns]

metabo_data = metabo_data.loc[:, all_samps]

# perform inchi strip
metabo_data['inchi_strip'] = [i.split("-")[0] for i in metabo_data.index.get_level_values(4)]

```

```python
chalm_inchi = pd.read_csv("./metabolomics/metabolomics_pathways/chalm_metabolite_inchi_withMissingGenes.csv",
                         sep = ",", index_col = 0)

chalm_inchi['inchi_strip'] = [i.split("-")[0] for i in chalm_inchi.inchi_key]
chalm_inchi.set_index('metabolite', inplace = True)
```

```python
print('getting sankey edges ....')
all_edges = []
ct_edges = []
chemical_edges = []


ct_subsys = []

for k in ['Proximal tubule']: #
    
    metabo_genes = all_gene_rxns[all_gene_rxns['index'].isin(allSigs[k])]
    gene_intersec = np.intersect1d(metabo_genes['index'].values.tolist(),
                     urine_logcpmtmm.index.get_level_values(0)) 
    metabo_genes = metabo_genes[metabo_genes['index'].isin(gene_intersec)]
    
    # drop duplicate (gene, subsystem)
    metabo_genes = metabo_genes[['index', 'subsystem']].drop_duplicates(keep = 'first')
    
    
    if metabo_genes.shape[0] > 0:
        
        for subsys in np.unique(metabo_genes.subsystem):
            num_genes_this_subsys = metabo_genes[metabo_genes.subsystem == subsys].shape[0]
            
            if num_genes_this_subsys > 1:
                ct_res = [(k, subsys, num_genes_this_subsys)]
                ct_edges += ct_res
                all_edges += ct_res
                
                
                #print(k, subsys, num_genes_this_subsys, 
                #          metabo_genes[metabo_genes.subsystem == subsys]['index'].values.tolist())
                ct_subsys += [subsys]
            
for subsys in np.unique(chem_class.subsystem):

    #for metabo_class in np.unique(chem_class.Class):
    for metabo_class in np.unique(chem_class['Parent Level 1']):
        subsys_subset = chem_class[chem_class.subsystem == subsys]
        subsys_subset = subsys_subset[subsys_subset['Parent Level 1'] == metabo_class]
            
        if subsys_subset.shape[0] > 0:
            
            metabo_res =  [(subsys, metabo_class, subsys_subset.shape[0])]
            all_edges += metabo_res
            chemical_edges += metabo_res
            
            #if subsys in ct_subsys:
            #    print(k, subsys, subsys_subset.metNames.values.tolist())
print('all sankey edges acquired.')
```

```python
# update the edge list 
gene_thresh = 2
metab_thresh = 3
ct_edges = [i for i in ct_edges if i[2] >= gene_thresh]
chemical_edges = [i for i in chemical_edges if i[2] >= metab_thresh]
```

```python
# get subsystems from cell types not represented on chemical side (and vice versa)
ct_subsys = [i[1] for i in ct_edges]
chem_subsys = [i[0] for i in chemical_edges]

ct_subsys_not_chemical = np.setdiff1d(ct_subsys, chem_subsys)
chemical_subsys_not_ct = np.setdiff1d(chem_subsys, ct_subsys)
shared_subsys = np.intersect1d(ct_subsys, chem_subsys)
```

```python
# this is the cmap for the sankey
meta_dict = {}
sankey_cmap = {}

cell_lst = ['Proximal tubule']

sankey_cmap['Proximal tubule'] = "#ffd000"

# subsystems, cell types, chemical families

subsys = np.unique(all_gene_rxns.subsystem)

celltypes = cell_lst

chemical_classes = np.unique(chem_class['Parent Level 1'])

num_unique_subsys = len(subsys)
num_celltypes = len(celltypes)

#for i, subsys in enumerate(subsys):
for i, subsys_i in enumerate(subsys):
    meta_dict[subsys_i] = i
    sankey_cmap[subsys_i] = colors[i]
    #meta_dict[subsys] = i
    #sankey_cmap[subsys] = colors[i]


for i, ct in enumerate(celltypes):
    meta_dict[ct] = i + num_unique_subsys


chem_class_dict = pd.read_pickle("chemicalclass_colors.pickle")
chem_class_dict = pd.read_pickle("20240111_chemicalclass_colors.pickle")


for i, chemical_class in enumerate(chemical_classes):
    meta_dict[chemical_class] = i + num_unique_subsys + num_celltypes
    
    if chemical_class in chem_class_dict.keys():
        sankey_cmap[chemical_class] = chem_class_dict[chemical_class]
    else:

        sankey_cmap[chemical_class] = colors[i]
```

```python
# create variable dummy_edges for facilitating flow
dummy_edges = []

for i in chemical_subsys_not_ct:
    dummy_edges += [('dummy_ct', i, 1)]
for i in ct_subsys_not_chemical:
    dummy_edges += [(i, 'dummy_chemical',1)]
```

```python
all_edges = chemical_edges + ct_edges + dummy_edges
```

```python
final_edges = []
for i in chemical_edges:
    if i[0] in shared_subsys:
        final_edges += [i]
for j in ct_edges:
    if j[1] in shared_subsys:
        final_edges += [j]
        
all_edges = final_edges 
```

```python
#hv.help(hv.Dimension)
##value_dim = hv.Dimension('Percentage', unit='%')
sankey = hv.Sankey(all_edges,
                   ['subsys', 'chemical_class']) #, vdims=value_dim)


sankey.opts(width = 500, height = 750,
            cmap = sankey_cmap,
            edge_color = dim('subsys').str(),
            node_color = dim('index').str())
```
